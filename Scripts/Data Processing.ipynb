```python
import pandas as pd
import numpy as np
import os
print(os.getcwd())
```

```python
# load the dataset
data = pd.read_csv("../Data/spam.csv", encoding="ISO-8859-1")
data.head()
```

```python
# remove extra columns
data = data.dropna(axis=1, how='any')

# rename existing columns
data.rename(columns = {"v1": "Classification", "v2": "Message"}, inplace = True)

data.head(5)
```

```python
# remove NAs
print(data.isna().any())

data.head()

# There are no NAs in this dataset.
```

```python
# Convert to 0s and 1s. 0 will be ham and 1 will be spam.
data["Classification"] = data["Classification"].replace({"ham": 0, "spam": 1})

data.head(15) # Taking a look to make sure it replaced ham and spam correctly.
```

```python
# remove non-ASCII characters and fix decoding issues
data["Message"] = (
    data["Message"]
    .str.encode("cp1252", errors="ignore")
    .str.decode("utf-8", errors="ignore")
)

data.head(25)
```

```python
# save as new csv file to the "Data" folder
data.to_csv("../Data/SMS_Processed.csv", index = False) 
```
