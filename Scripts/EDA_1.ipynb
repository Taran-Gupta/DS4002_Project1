```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset
data = pd.read_csv("../Data/spam.csv", encoding='ISO-8859-1')
```

```python
# Drop NAs 
data = data.dropna(axis = 1, how = "all")

# Calculate the percentage of messages that contain the word "money"
data['v2'].str.contains('money', case=False, na=False).sum()

money_count = data.loc[data['v1'] == 'spam', 'v2'].str.contains('money', case=False, na=False).mean() * 100

print("Money Percentage is " + str(money_count))

# Calculate the percentage of messages that contain the word "win"
win_count = data.loc[data['v1'] == 'spam', 'v2'].str.contains('win', case=False, na=False).mean() * 100

print("Win Percentage is " + str(win_count))

# Calculate the percentage of messages that contain the word "free"
free_count = data.loc[data['v1'] == 'spam', 'v2'].str.contains('free', case=False, na=False).mean() * 100

print("Free Percentage is " + str(free_count))

# Calculate the percentage of messages that contain the word "immediately"
immediately_count = data.loc[data['v1'] == 'spam', 'v2'].str.contains('immediately', case=False, na=False).mean() * 100

print("Immediately Percentage is " + str(immediately_count))
```

```python
# Plot the distribution of spam vs. ham messages
counts = data["v1"].value_counts()

counts.plot(kind = 'bar', color = ["blue", "orange"])
plt.title("Distribution of Spam vs. Ham Messages", fontsize = 15, weight = 'bold')
plt.xlabel("Message Type", fontsize = 12)
```

```python
# Calculate the length of each message and add it as a new column
data["Message Length"] = data["v2"].str.len()

# Create boxplot showing the distribution of message lengths for spam vs. ham messages (median, quartiles, outliers)
data.boxplot(column = "Message Length", by = "v1")

plt.title("Message Length Grouped by Message Type")
plt.xlabel("Message Type")

plt.suptitle("")
```

```python
# Continuously plot the density of message lengths for spam vs. ham messages
data[data['v1'] == "ham"]["Message Length"].plot(kind = "density")
data[data['v1'] == "spam"]["Message Length"].plot(kind = "density")
```
